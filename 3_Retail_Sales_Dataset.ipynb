```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('ggplot')
data = {
'Product_ID':['PR01','PR02','PR03','PR04','PR05','PR06','PR07','PR08','PR09','PR10',
              'PR11','PR12','PR13','PR14','PR15','PR16','PR17','PR18','PR19','PR20'],
'Category':['Grocery','Snacks','Dairy','Fruits','Vegetables',
            'Grocery','Snacks','Dairy','Fruits','Vegetables',
            'Grocery','Snacks','Dairy','Fruits','Vegetables',
            'Grocery','Snacks','Dairy','Fruits','Vegetables'],
'Price':[50,30,45,60,40,80,25,55,70,35,90,20,65,75,45,100,35,50,80,30],
'Quantity_Sold':[20,35,18,25,30,15,40,16,22,28,12,45,14,20,26,10,32,19,18,35],
'Discount':[5,3,2,5,4,6,3,2,5,4,6,2,3,5,4,7,3,2,6,3],
'Revenue':[1000,1050,810,1500,1200,1200,1000,880,1540,980,
           1080,900,910,1500,1170,1000,1120,950,1440,1050]
}
df=pd.DataFrame(data)
print("Retail Sales Dataset")
display(df)
print("Shape :",df.shape)
print("Rows :",df.shape[0])
print("Columns :",df.shape[1])
print("\nColumn Names")
print(df.columns)
print("\nData Types")
print(df.dtypes)
print("\nDataset Information")
df.info()
print("""

Purpose

This dataset helps analyze supermarket sales
using Product Category, Price, Quantity Sold,
Discount and Revenue.

EDA identifies the highest-selling products,
revenue patterns and customer purchasing trends.

""")
print("Missing Values")
print(df.isnull().sum())
df.fillna(method='ffill',inplace=True)
print("Duplicate Records :",df.duplicated().sum())
df.drop_duplicates(inplace=True)
display(df)
display(df.describe())
print("\nMean")
print(df.mean(numeric_only=True))
print("\nMedian")
print(df.median(numeric_only=True))
print("\nMinimum")
print(df.min(numeric_only=True))
print("\nMaximum")
print(df.max(numeric_only=True))
print("\nStandard Deviation")
print(df.std(numeric_only=True))
print(df['Category'].value_counts())
print(df['Category'].value_counts(normalize=True)*100)
display(df.corr(numeric_only=True))
plt.figure(figsize=(8,5))
sns.countplot(x='Category',data=df,palette='Set2')
plt.title("Products by Category")
plt.xlabel("Category")
plt.ylabel("Count")
plt.xticks(rotation=20)
plt.show()
plt.figure(figsize=(6,5))
plt.hist(df['Price'],bins=6,color='skyblue',edgecolor='black')
plt.title("Price Distribution")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.show()
plt.figure(figsize=(6,5))
plt.hist(df['Revenue'],bins=6,color='orange',edgecolor='black')
plt.title("Revenue Distribution")
plt.xlabel("Revenue")
plt.ylabel("Frequency")
plt.show()
plt.figure(figsize=(6,5))
sns.scatterplot(x='Price',
                y='Revenue',
                hue='Category',
                data=df,
                s=120)
plt.title("Price vs Revenue")
plt.show()
plt.figure(figsize=(6,5))
sns.scatterplot(x='Quantity_Sold',
                y='Revenue',
                hue='Category',
                data=df,
                s=120)
plt.title("Quantity Sold vs Revenue")
plt.show()
plt.figure(figsize=(6,5))
sns.boxplot(y='Price',data=df,color='lightgreen')
plt.title("Price Box Plot")
plt.show()
plt.figure(figsize=(6,5))
sns.boxplot(y='Revenue',data=df,color='pink')
plt.title("Revenue Box Plot")
plt.show()
plt.figure(figsize=(7,7))
df['Category'].value_counts().plot(
kind='pie',
autopct='%1.1f%%'
)
plt.title("Category Distribution")
plt.ylabel("")
plt.show()
plt.figure(figsize=(7,5))
sns.heatmap(df.corr(numeric_only=True),
annot=True,
cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
plt.figure(figsize=(8,5))
plt.plot(df['Revenue'],marker='o',label='Revenue')
plt.plot(df['Quantity_Sold'],marker='s',label='Quantity Sold')
plt.legend()
plt.title("Revenue and Quantity Sold Trend")
plt.xlabel("Products")
plt.ylabel("Values")
plt.show()
print("""

OBSERVATIONS

1. Dataset contains 20 products.

2. No missing values are present.

3. No duplicate records are found.

4. Fruits category generates the highest revenue.

5. Snacks have the highest quantity sold.

6. Revenue generally increases with price and quantity sold.

7. Very few outliers are present.

""")
print("""

KEY FINDINGS

• Fruits contribute the highest revenue.

• Snacks have maximum sales volume.

• Grocery products have higher prices.

• Revenue is positively related to Price and Quantity Sold.

• Discount has only a small effect on revenue.

""")
print("""

CONCLUSION

The Retail Sales dataset shows that product
category, price and quantity sold directly
influence revenue generation.

Fruits and Grocery products generate higher
income, while Snacks contribute the largest
sales volume.

The dataset is clean and suitable for
further sales forecasting and business analysis.

""")
print("""

DECISION MAKING

This analysis helps the supermarket to

• Identify high-performing product categories.

• Improve inventory management.

• Optimize pricing strategies.

• Plan discount offers effectively.

• Increase revenue through data-driven decisions.

""")
```
