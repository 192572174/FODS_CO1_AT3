```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('ggplot')
data = {
'Customer_ID':['C01','C02','C03','C04','C05','C06','C07','C08','C09','C10',
               'C11','C12','C13','C14','C15','C16','C17','C18','C19','C20'],
'Income':[45000,25000,60000,22000,50000,30000,70000,28000,55000,24000,
          65000,35000,48000,26000,75000,32000,52000,27000,68000,23000],
'Credit_Score':[720,580,750,560,710,600,780,590,730,570,
                760,620,700,585,790,610,725,575,770,555],
'Loan_Amount':[200000,150000,300000,120000,250000,180000,350000,160000,270000,130000,
               320000,200000,240000,140000,400000,190000,260000,150000,330000,125000],
'Employment_Type':['Salaried','Self-employed','Salaried','Unemployed','Salaried',
                   'Self-employed','Salaried','Self-employed','Salaried','Unemployed',
                   'Salaried','Self-employed','Salaried','Unemployed','Salaried',
                   'Self-employed','Salaried','Self-employed','Salaried','Unemployed'],
'Loan_Status':['Approved','Rejected','Approved','Rejected','Approved',
               'Rejected','Approved','Rejected','Approved','Rejected',
               'Approved','Rejected','Approved','Rejected','Approved',
               'Rejected','Approved','Rejected','Approved','Rejected']
}
df = pd.DataFrame(data)
print("Banking Loan Approval Dataset")
display(df)
print("Shape :", df.shape)
print("Rows :", df.shape[0])
print("Columns :", df.shape[1])
print("\nColumn Names")
print(df.columns)
print("\nData Types")
print(df.dtypes)
print("\nDataset Information")
df.info()
print("""

Purpose

This dataset is used to analyze loan approval
based on Income, Credit Score, Loan Amount
and Employment Type.

EDA helps identify the important factors
that influence loan approval decisions.

""")
print("\nMissing Values")
print(df.isnull().sum())
df.fillna(method='ffill', inplace=True)
print("\nDuplicate Records :", df.duplicated().sum())
df.drop_duplicates(inplace=True)
df['Loan_Code'] = df['Loan_Status'].map({'Approved':1,'Rejected':0})
display(df)
display(df.describe())
print("\nMean")
print(df.mean(numeric_only=True))
print("\nMedian")
print(df.median(numeric_only=True))
print("\nMinimum")
print(df.min(numeric_only=True))
print("\nMaximum")
print(df.max(numeric_only=True))
print("\nStandard Deviation")
print(df.std(numeric_only=True))
print(df['Employment_Type'].value_counts())
print(df['Loan_Status'].value_counts())
print(df['Loan_Status'].value_counts(normalize=True)*100)
display(df.corr(numeric_only=True))
plt.figure(figsize=(7,5))
sns.countplot(x='Loan_Status', data=df, palette='Set2')
plt.title("Loan Approval Status")
plt.xlabel("Loan Status")
plt.ylabel("Count")
plt.show()
plt.figure(figsize=(8,5))
sns.countplot(x='Employment_Type', hue='Loan_Status', data=df)
plt.title("Employment Type vs Loan Status")
plt.xlabel("Employment Type")
plt.ylabel("Count")
plt.show()
plt.figure(figsize=(6,5))
plt.hist(df['Income'], bins=6, color='skyblue', edgecolor='black')
plt.title("Income Distribution")
plt.xlabel("Income")
plt.ylabel("Frequency")
plt.show()
plt.figure(figsize=(6,5))
plt.hist(df['Credit_Score'], bins=6, color='orange', edgecolor='black')
plt.title("Credit Score Distribution")
plt.xlabel("Credit Score")
plt.ylabel("Frequency")
plt.show()
plt.figure(figsize=(6,5))
sns.scatterplot(x='Income',
                y='Credit_Score',
                hue='Loan_Status',
                data=df,
                s=120)
plt.title("Income vs Credit Score")
plt.show()
plt.figure(figsize=(6,5))
sns.scatterplot(x='Loan_Amount',
                y='Credit_Score',
                hue='Loan_Status',
                data=df,
                s=120)
plt.title("Loan Amount vs Credit Score")
plt.show()
plt.figure(figsize=(6,5))
sns.boxplot(y='Income', data=df, color='lightgreen')
plt.title("Income Box Plot")
plt.show()
plt.figure(figsize=(6,5))
sns.boxplot(y='Credit_Score', data=df, color='pink')
plt.title("Credit Score Box Plot")
plt.show()
plt.figure(figsize=(6,6))
df['Loan_Status'].value_counts().plot(
    kind='pie',
    autopct='%1.1f%%',
    colors=['lightgreen','salmon']
)
plt.title("Loan Approval Percentage")
plt.ylabel("")
plt.show()
plt.figure(figsize=(7,5))
sns.heatmap(df.corr(numeric_only=True),
            annot=True,
            cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
plt.figure(figsize=(8,5))
plt.plot(df['Income'], marker='o', label='Income')
plt.plot(df['Credit_Score'], marker='s', label='Credit Score')
plt.title("Income and Credit Score Trend")
plt.xlabel("Customers")
plt.ylabel("Values")
plt.legend()
plt.show()
print("""

OBSERVATIONS

1. Dataset contains 20 customer records.

2. No missing values are present.

3. No duplicate records are found.

4. Customers with higher income generally have
higher credit scores.

5. Salaried customers receive more loan approvals.

6. Approved loans are associated with
higher credit scores and income.

7. Very few outliers are observed.

""")
print("""

KEY FINDINGS

• Income strongly influences loan approval.

• Higher credit scores increase approval chances.

• Salaried employees have a higher approval rate.

• Loan amount depends on customer income.

• Credit score and income show a positive relationship.

""")

print("""

CONCLUSION

The Banking Loan dataset indicates that
Income, Credit Score and Employment Type
are the major factors affecting loan approval.

Customers with higher income and better
credit scores are more likely to receive
loan approval.

The dataset is clean and suitable for
predictive modeling and banking analysis.

""")
print("""

DECISION MAKING

This analysis helps banks to

• Identify eligible loan applicants.

• Reduce loan default risk.

• Improve credit assessment.

• Make faster loan approval decisions.

• Develop fair and data-driven lending policies.

""")
```
