```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('ggplot')
data = {
    'Patient_ID':['P01','P02','P03','P04','P05','P06','P07','P08','P09','P10',
                  'P11','P12','P13','P14','P15','P16','P17','P18','P19','P20'],
    'Age':[45,32,55,28,60,40,52,35,48,30,62,38,50,27,58,42,46,34,53,29],
    'Sugar_Level':[145,98,180,90,200,120,165,105,155,95,
                   210,115,170,88,190,130,150,100,175,92],
    'BP':[130,118,145,110,150,125,140,120,135,115,
          155,122,142,108,148,128,132,116,144,112],
    'BMI':[28,23,31,22,34,26,30,24,29,23,
           35,25,30,21,33,27,28,24,31,22],
    'Disease_Status':['Yes','No','Yes','No','Yes','No','Yes','No','Yes','No',
                      'Yes','No','Yes','No','Yes','No','Yes','No','Yes','No']
}
df = pd.DataFrame(data)
print("Healthcare Analytics Dataset")
display(df)
print("Shape:",df.shape)
print("Rows:",df.shape[0])
print("Columns:",df.shape[1])
print("\nColumn Names")
print(df.columns)
print("\nData Types")
print(df.dtypes)
print("\nInformation")
df.info()
print("""
Purpose:

To analyze patient health conditions using
Age, Sugar Level, Blood Pressure and BMI.

The analysis identifies the factors associated
with disease occurrence.
""")
print(df.isnull().sum())
df.fillna(method='ffill',inplace=True)
print("Duplicate Records :",df.duplicated().sum())
df.drop_duplicates(inplace=True)
df['Disease_Code']=df['Disease_Status'].map({'Yes':1,'No':0})
display(df)
display(df.describe())
print("\nMean")
print(df.mean(numeric_only=True))
print("\nMedian")
print(df.median(numeric_only=True))
print("\nMinimum")
print(df.min(numeric_only=True))
print("\nMaximum")
print(df.max(numeric_only=True))
print("\nStandard Deviation")
print(df.std(numeric_only=True))
print(df['Disease_Status'].value_counts())
print(df['Disease_Status'].value_counts(normalize=True)*100)
display(df.corr(numeric_only=True))
plt.figure(figsize=(6,5))
sns.countplot(x='Disease_Status',data=df,palette='Set2')
plt.title("Disease Status Distribution")
plt.xlabel("Disease Status")
plt.ylabel("Count")
plt.show()
plt.figure(figsize=(6,5))
plt.hist(df['Age'],bins=6,color='skyblue',edgecolor='black')
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
plt.figure(figsize=(6,5))
plt.hist(df['Sugar_Level'],bins=6,color='orange',edgecolor='black')
plt.title("Blood Sugar Distribution")
plt.xlabel("Sugar Level")
plt.ylabel("Frequency")
plt.show()
plt.figure(figsize=(6,5))
sns.scatterplot(x='Sugar_Level',
                y='BMI',
                hue='Disease_Status',
                data=df,
                s=120)
plt.title("Sugar Level vs BMI")
plt.show()
plt.figure(figsize=(6,5))
sns.scatterplot(x='Age',
                y='BP',
                hue='Disease_Status',
                data=df,
                s=120)
plt.title("Age vs Blood Pressure")
plt.show()
plt.figure(figsize=(6,5))
sns.boxplot(y='Sugar_Level',data=df,color='lightgreen')
plt.title("Blood Sugar Box Plot")
plt.show()
plt.figure(figsize=(6,5))
sns.boxplot(y='BMI',data=df,color='pink')
plt.title("BMI Box Plot")
plt.show()
plt.figure(figsize=(6,6))
df['Disease_Status'].value_counts().plot(
kind='pie',
autopct='%1.1f%%',
colors=['tomato','lightgreen']
)
plt.title("Disease Status")
plt.ylabel("")
plt.show()
plt.figure(figsize=(7,5))
sns.heatmap(df.corr(numeric_only=True),
annot=True,
cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
plt.figure(figsize=(8,5))
plt.plot(df['Sugar_Level'],marker='o',label='Sugar Level')
plt.plot(df['BP'],marker='s',label='Blood Pressure')
plt.legend()
plt.title("Sugar Level and Blood Pressure Trend")
plt.xlabel("Patients")
plt.ylabel("Values")
plt.show()
print("""

OBSERVATIONS

1. Dataset contains 20 patient records.

2. No missing values were found.

3. No duplicate records exist.

4. Patients with high sugar levels generally have
higher BMI and Blood Pressure.

5. Disease-positive patients are mostly above
45 years of age.

6. Sugar Level shows a strong relationship
with Disease Status.

""")
print("""

KEY FINDINGS

• High Blood Sugar increases disease risk.

• Older patients have greater disease occurrence.

• High BMI is associated with disease.

• Blood Pressure is higher in diseased patients.

• Health indicators are positively correlated.

""")
print("""

CONCLUSION

EDA shows that Age, Sugar Level,
Blood Pressure and BMI play an
important role in predicting disease.

The dataset is clean without
missing or duplicate values and
is suitable for further predictive analysis.

""")
print("""

DECISION MAKING

The hospital can use this analysis to

• Identify high-risk patients.

• Recommend regular health screening.

• Monitor diabetic patients.

• Plan preventive healthcare programs.

• Improve clinical decision-making.

""")
```
